<div dir=rtl id="start" style="direction:rtl;line-height:200%;">
	<font face="vazir" size=3>
با استفاده از تابع <code>filter</code> و انجام گروه بندی، وب‌سایت‌هایی را انتخاب کنید که <b>کلید واژه‌ی مربوط به آن‌ها، بیشتر از ۱۴۰۰ وب‌سایت را شامل شود.</b>
        <br>
        <br>
راهنمایی: ابتدا تابع <code>groupby</code> را برای کلید‌واژه‌ها اعمال کرده و سپس از از تابع <code>len</code> برای بررسی تعداد اعضا استفاده کنید. توجه کنید <b> لازم نیست </b>گروه‌بندی براساس هر عضو داخل آرایه کلیدواژه‌ها انجام شود.
    </font>
</div>


```python
filtered_df = df.groupby("keywords").filter(lambda x : len(x) >1400) 
```

<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
قسمت دوم
    </font>
</h2>



<div dir=rtl id="start" style="direction:rtl;line-height:200%;">
	<font face="vazir" size=3>
می‌خواهیم ستونی برای نشان دادن درصد شهرت هر وب‌سایت با توجه به کلید‌واژه‌های آن اضافه کنیم.
<br>
برای این کار، ستونی با نام <code>'popularity'</code> را به دیتافریم <code>filtered_df</code> اضافه کنید که نسبت تعداد ویزیت هر سایت به مجموع تعداد ویزیت تمامی سایت‌های آن کلید‌واژه را نشان دهد.
<br>
<br>
راهنمایی: بعد از انجام گروه‌بندی بر اساس <code>keywords</code>، روی ستون <code>visit count</code> تابع <code>transfrom</code> را اعمال کرده و نسبت ویزیت هر سایت به مجموع ویزیت‌های آن گروه کلید واژه را محسابه و در ۱۰۰ ضرب کنید.
    </font>
</div>


```python
filtered_df['popularity']  = filtered_df.groupby('keywords')['visit count'].transform(lambda x: (x/x.sum()*100))
filtered_df
```

```python
import zlib
import zipfile


def compress(file_names):
    print("File Paths:")
    print(file_names)
    compression = zipfile.ZIP_DEFLATED
    with zipfile.ZipFile("result.zip", mode="w") as zf:
        for file_name in file_names:
            zf.write('./' + file_name, file_name, compress_type=compression)


filtered_df.to_csv('filtered_df.csv',index = True)

file_names = ["filtered_df.csv", "yektanet.ipynb"]
compress(file_names)
```
